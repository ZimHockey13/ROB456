```python
import numpy as np
```

```python
x_door = {
    "Open": 0.6,
    "Closed": 0.4
}

# end state | action | initial state
action_table = {
    "Open|Open|Open": 1,
    "Closed|Open|Open": 0,
    "Open|Open|Closed": 0.8,
    "Closed|Open|Closed": 0.2,
    "Closed|Closed|Open": 0.7,
    "Open|Closed|Open": 0.3,
    "Closed|Closed|Closed": 1,
    "Open|Closed|Closed": 0
}
```

```python
def do_action(x_door, action_table, action_chosen):
    new_probs = {
        "Open": 0,
        "Closed": 0
    }

    for key1, value1 in x_door.items():
        for key2, value2 in x_door.items():
            new_probs[key1] += action_table[f"{key1}|{action_chosen}|{key2}"]*value2

    return new_probs
```

```python
print(f"action to open: {do_action(x_door, action_table, "Open")}")
print(f"action to close: {do_action(x_door, action_table, "Closed")}")
```
